Множество приема запросов на предоставление услуги $i$-типа, $i$=1,2:
\begin{equation}

S_i = \bar{{B_i}} = {X} \backslash {B_i} = \{0,1,,,.,C-1\}

Система уравнений глобального баланса (СУГБ):

\begin{equation}
\tag{1.4}
\begin{cases}
    (\lambda_1 + \lambda_2) p_0 = μ p_1), \\
    (\lambda_1 + \lambda_2 + n\mu)p_n = (\lambda_1 + \lambda_2)p_{n-1} + (n+1)\mu p_{n+1}, n = \overline{{1,C-1}}, \\
    C\mu p_C = (\lambda_1 + \lambda_2)p_{C-1}
\end{cases}
\end{equation}

Система уравнений локального баланса (СУЛБ):
\begin{equation}
\tag{1.5}
(\lambda_1+\lambda_2)p_{n-1} = n \mu p_n ,    n = \overline{1, C}
\end{equation}

Стационарное распределение вероятностей состояний системы:
\begin{equation}
\tag{1.6}
p_n = \left( \sum_{i=0}^{C}{\frac{(\rho_1+\rho_2)^{i}}{i!}}\right)^{-1}. \frac{\left(\rho_1+\rho_2\right)^{n}}{n!}, n = \overline{0, C}
\end{equation}

#### Доказательство:

Используя СУЛБ, найдем стационарное распределение вероятностей состояний системы $p_n, n=\overline{1, C}:$

$$
p_n [C] = p_n = p_{n-1}. \frac{\lambda_1+\lambda_2}{n\mu} = p_{n_1}. \frac{\rho_1+\rho_2}{n} = ... = p_0. \frac{(\rho_1+\rho_2)^{n}}{n!}, n = \overline{1, C}
$$


Для нахождения вероятности $p_0$ воспользуемся условием нормировки $\sum_{n=0}^{C} p_n = 1$:

$$
p_0 = p_0 [C] = \left(\sum_{n = 0}^{C} {\frac{(\rho_1+\rho_2)^{n}}{n!}}\right)^{-1}
$$

Основные вероятностные характеристики (ВХ) модели:

## **2. Численный анализ**

Создание программы, реализующей расчет распределения вероятностей,
вероятности блокировки, среднего количества обслуженных запросов для любых значений исходных данных. Программа должна выводить на экран:
* значение распределения вероятностей,
* значения вероятностей блокировки,
* значение среднего числа заявок.

Код написан на языке Python:

```python
import math #for power and factorial function
import matplotlib.pyplot as plt #for data visualization - to plot graph
import numpy as np #for array computing
```

Определение функции $p_0$ для вычисления вероятности того, что система находится в исходном состоянии:

```python
def p_0(C, rho):
    # rho = rho1+rho2
    probab = 0
    if C != 0:
        sum = 0
        for n in range(C):
            sum += math.pow(rho, n)/math.factorial(n)
        probab = 1/sum
    return probab
```

Определение функции $p_n$ для вычисления вероятности того, что система находится в n состоянии:

```python
def p_n(C, n, rho):
    p_nC = ((math.pow(rho, n) / math.factorial(n))) * p_0(C=C, rho=rho)
    return p_nC
```

Определения функция $p_C$ для вычисления вероятности блокировки по нагрузке и по времени $C_1=C_2=E$, в которой $n=C$

```python
def p_C(C, rho):
    # compute E, C1, C2
    return p_n(C=C, n=C, rho=rho)
```

Определение функций **mean_request** для вычисления среднего число запросов в системе $\bar N$:

```python
def mean_request(C, rho):
    expecation = 0
    for n in range(C):
        expecation += n * p_n(C=C, n=n, rho=rho)
    return expecation
```

Для расчета основных вероятностных характеристик модели были взяты следующие параметры:

\begin{equation}
 C = 100, μ_1=μ_2=μ=0.8, λ_1=40, λ_2=60
\end{equation}
Заметить что:
$$
 ρ_1=λ_1/μ_1 \\
 ρ_2=λ_2/μ_2
$$
$ρ=ρ_1+ρ_2$ может варьироваться от 0 до $λ_1/μ_1+λ_2/μ_2=125$

```python
C = 100
mu = 0.8
lambda_1 = 60
lambda_2 = 40
```

Проверка значения стационарного распределения вероятностей в состоянии $n = 30, 33, ..., 50$ с увеличением на $3$:

```python
Rho = (lambda_1 + lambda_2) / mu
for n in np.arange(30, 50, 3):
    print("Распределения вероятностей того, что система находится в {}-ый состоянии, равна {}".format(n, p_n(C=C, n=n, rho=Rho)))
```

```python
array_p_n = np.array([], 'float64')
for n in range(C):
    array_p_n = np.append(array_p_n, p_n(C=C, n=n, rho=Rho))
print("Распределения вероятностей нахождения системы в каждом состоянии n: ", array_p_n)
```

Вероятность блокировки по времени $E$:

```python
# just apply the function p_C defined above
E = p_C(C=C, rho=Rho)
print("Вероятность блокировки по времени E =", E)
```

Вероятность блокировки по вызовам $B_i$:

```python
# use the relation between B1, B2 with E as given in lecture
B_1 = (lambda_1 / (lambda_1 + lambda_2)) * E
B_2 = (lambda_2 / (lambda_1 + lambda_2)) * E
print("B1 = {}, B2 = {}".format(B_1, B_2))
```

Среднее число обслуживаемых в системе запросов:

```python
N = mean_request(C=C, rho=Rho)
print("Среднее число N обслуживаемых в системе запросов = ", N)
```

* Построение график график зависимости вероятности блокировки от интенсивности поступления запросов на обслуживание

```python
# Incrementation of lambda between 0 and C, with incremental step of 0.4
#Увеличение значения lambda от 0 до C с шагом увеличения 0,4
lambda__ = np.arange(0, lambda_1+lambda_2, 0.4)

# array_E of p_C when lmabda change
rho = lambda__ / mu
array_E = []
for ro in rho:
    array_E.append(p_C(C=C, rho=ro))
plt.plot(lambda__, array_E, color='blue', label='$E=C_i$')
plt.title("Вероятность блокировки по времени и вероятность блокировки по нагрузке в зависимости от интенсивности запросов")
plt.xlabel('Интенсивность поступления запросов $\lambda_1+\lambda_2$')
plt.ylabel('Значение вероятность')
plt.legend()
plt.show()
```

```python
# array_E of p_C when lambda change
B_1_array = []
for lam, E in zip(lambda__, array_E):
    B_1_array.append(lam / (lam + lambda_2))
B_2_array = []
for lam, E in zip(lambda__, array_E):
    B_2_array.append(lambda_2 / (lam + lambda_2))
plt.plot(lambda__, B_1_array, color= 'blue', label="$B_1$")
plt.plot(lambda__, B_2_array, color= 'green', label="$B_2$")
plt.title("Вероятность блокировки по вызовам в зависимости от интенсивности запросов")
plt.xlabel('Интенсивность поступления запросов $\lambda_1+\lambda_2$')
plt.ylabel('Вероятность блокировки')
plt.legend()
plt.show()
```

* Построение график среднего числа обслуживаемых запросов от интенсивности поступления запросов:

```python
mean_request_array = []
rho_ = lambda__ / mu
for rho in rho_:
    mean_request_array.append(mean_request(C=C, rho=rho))

plt.plot(rho_, mean_request_array, color="green", label = 'N')
plt.title("Среднее число обслуживаемых в зависимости от интенсивности запросов")
plt.xlabel('Интенсивность поступления запросов $\lambda_1 + \lambda_2$')
plt.ylabel('Среднее число запросов')
plt.legend()
plt.show()
```
