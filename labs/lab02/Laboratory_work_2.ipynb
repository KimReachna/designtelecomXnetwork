# An incomplete two-service Erlang model with the same service intensities and reserved capacity

Consider a network link with a capacity of C . Let the network users be provided with two types of services. Requests for services are Poisson flows with intensities $\rho_1$, $\rho_2$, . The average time to service requests of each type $\mu_1 ^ {-1}, \mu_2 ^ {-1}$, respectively. Consider the case of $μ_1=\mu_2=\mu$.

Part of the cell's bandwidth is reserved for servicing requests for the provision of type 1 or type 2 services. The remaining part of the bandwidth is fully available for requests for both types of services. Suppose that a fully accessible container is filled first.



* $C$ - peak cell throughput;
* $g$ - the fully accessible part of the cell's bandwidth;
* $C-g$ - the bandwidth reserved for servicing requests for the provision of type 1 or type 2 services;
* $λ_1,λ_2$ - the intensity of receipt of requests for the provision of services of the 1st, 2nd type [запросов/ед.вр.];
* $μ^{-1}$ - average service time of a request for the provision of a type 1, type 2 service [запросов/ед.вр.];
* $ρ_1,ρ_2$ - the intensity of the proposed load created by requests for the provision of type 1, type 2 services;
* $X(t)$ - the number of requests served in the system at time $t , t ≥ 0$ (random process describing the functioning of the system at time $t , t ≥ 0$);
* $X$ - system state space;
* $n$ - the number of requests served in the system;
* $B_1, B_2$ - set of blocking requests for the provision of services of the 1st, 2nd type;
* $S_1,S_2$ - the set of receiving requests for the provision of services of the 1st, 2nd type.

Let's assume that part of the bandwidth of cell $C − g$ is reserved for servicing requests for the provision of type 1 services.

**Set of blocking requests for the provision of services of the $i$-type**, $i$=1,2:
\begin{equation}
\tag{2.2}
B_1={C},
\end{equation}

\begin{equation}
\tag{2.3}
B_2={g,g+1,...,C}
\end{equation}

**The set of receiving requests for the provision of services of the $i$-type**, $i$=1,2:
\begin{equation}
\tag{2.4}
S_1 = \bar{{B_1}} = {X} \backslash {B_1} = \{0,...,C-1\}
\end{equation}

\begin{equation}
\tag{2.5}
S_2 = \bar{{B_2}} = {X} \backslash {B_2} = \{0,...,C_1-1\}
\end{equation}


**System of Global Balance Equations:**

\begin{equation}
\tag{2.6}
\begin{cases}
    (\lambda_1 + \lambda_2) p_0 = μ p_1, \\
    (\lambda_1 + \lambda_2 + n\mu)p_n = (\lambda_1 + \lambda_2)p_{n-1} + (n+1)\mu p_{n+1}, n = \overline{{1,g-1}}, \\
    (\lambda_1 + g\mu)p_g = (\lambda_1 + \lambda_2)p_{g-1} + (g+1)\mu p_{g+1}, \\
    (\lambda_1 + n\mu)p_n = \lambda_1 p_{n-1} + (n+1)\mu p_{n+1}, n = \overline{{g+1,C-1}}, \\
    C\mu p_C = \lambda_1 p_{C-1}
\end{cases}
\end{equation}

**System of Local balance equations:**

\begin{equation}
\tag{2.7}
\begin{cases}
  (\lambda_1+\lambda_2)p_{n-1} = n \mu p_n ,    n = \overline{1, g}, \\
  lanbda_1 p_{n-1} = n\mu p_n, n = \overline{{g+1,C}}
\end{cases}
\end{equation}

Обозначим:
\begin{equation}
\rho_1 = \frac{\lambda_1}{\mu},\\
\rho_2 = \frac{\lambda_2}{\mu}
\end{equation}

**Stationary probability distribution of system states:**

\begin{equation}
\tag{2.8}
p_n =
\begin{cases}
p_0 . \frac{(\rho_1+\rho_2)^{n}}{n!}, n=\overline{{1,g}} \\
p_0 . \frac{((\rho_1+\rho_2)^{g}) . ((\rho_1)^{n-g})}{n!}, n=\overline{{g+1,C}}
\end{cases}
\end{equation}

\begin{equation}
\tag{2.9}
p_0=\left(\sum_{n=1}^{g} {\frac{(\rho_1+\rho_2)^{n}}{n!}} + \sum_{n=g+1}^{C} {\frac{(\rho_1+\rho_2)^{g} . (\rho_1)^{n-g}}{n!}}\right)
\end{equation}


**Basic probabilistic characteristics of the model:**

* The probability of blocking by time $E_1$ of a request for the provision of a type 1-st service

\begin{equation}
\tag{2.10}
E_1 = \sum_{n\in B_1} p_n = p_C
\end{equation}

* The probability of blocking by time $E_1$ of a request for the provision of a type 2-nd service

\begin{equation}
\tag{2.11}
E_2 = \sum_{n\in B_2} p_n = p_g + p_{g+1} + ...+ p_C = \sum_{n=g}^{C} p_n
\end{equation}

* Average number of $\bar{N}$ requests served in the system:
\begin{equation}
\tag{2.12}
\bar{N} = \sum_{n\in X} np_n .
\end{equation}

```python
import numpy as np
import math
```

```python
def probability_0(rho1, rho2, g, C):
    '''
    Определите функцию для вычисления вероятности того, что система находится в состоянии 0, то есть все серверы системы бездействуют или свободны.
    '''
    sum_  = 0
    for n in np.arange(C+1):
        if n <= g:
            sum_ += math.pow(rho1 + rho2, n) / math.factorial(n)
        else:
            sum_ += math.pow(rho1 + rho2, g) * math.pow(rho1, n - g) / math.factorial(n)
    return 1 / sum_

def probability_n(n, rho1, rho2, g, C):
    '''
    Определенная функция для вычисления вероятности стационарности системы, если система стационарна в состоянии n, где 1 <= n <= C.
    '''
    p_n = 0 # initialize p_n
    p_0 = probability_0(rho1=rho1, rho2=rho2, g=g, C=C)
    if n == 0:
        # print("n = 0")
        p_n = p_0
    elif n < g + 1:
        # print("1 to g")
        p_n = p_0 * math.pow(rho1 + rho2, n) / math.factorial(n)
    elif n < C +1:
        # print("g+1 to C")
        p_n = (p_0 * math.pow(rho1 + rho2, g) * math.pow(rho1, n - g)) / math.factorial(n)
    return p_n
```

```python
C = 50
g = 35
lambda1 = 30
lambda2 = 20
mu = 0.8
rho1 = lambda1/mu
rho2 = lambda2/mu
```

## task 1 and task 2

```python
p_0 = probability_0(rho1=rho1, rho2=rho2, g=g, C=C)
print("Распределения вероятностей того, что система находится в состоянии 0-го: {}".format(p_0))
# this is valide for the above system configuration only.
```

```python
# print values of probability of systems and compare total to 1
probability_n_array = []
for n in np.arange(C+1):
    probability_n_array.append(probability_n(n=n, rho1=rho1, rho2=rho2, g=g, C=C))
print("Распределения вероятностей для каждого n:", probability_n_array)
print("probability_n_array размер: ", len(probability_n_array))
print("Сумма всех вероятностей: ", sum(probability_n_array))
```

```python
# blocking probability of service of type-1 if when just the probability_n when n = C
def p_C(C, rho1, rho2, g):
    return probability_n(n=C, rho1=rho1, rho2=rho2, g=g, C=C)
E1 = p_C(C=C, rho1=rho1, rho2=rho2, g=g)
print("Вероятность блокировки по времени E1 = p_C = {}".format(E1))
```

```python
# blocking proability of service of type-2 if when system is at least in state g and onward to C
def sum_pngc(rho1, rho2, g, C):
    return sum([probability_n(n=i, rho1=rho1, rho2=rho2, g=g, C=C) for i in np.arange(g, C+1)])
E2 = sum_pngc(rho1=rho1, rho2=rho2, g=g, C=C)
print('Вероятность блокировки по времени E_2 = {}'.format(E2))
```

```python
# calculate average number of service being served in system which is just the expecation of system station probability
def average_N_floored(rho1, rho2, g, C):
    return math.floor(sum([n * probability_n(n=n, rho1=rho1, rho2=rho2, g=g, C=C) for n in np.arange(C+1)]))

def average_N(rho1, rho2, g, C):
    return sum([n * probability_n(n=n, rho1=rho1, rho2=rho2, g=g, C=C) for n in np.arange(C+1)])

# compute the the number of N for the current system configuration

N = average_N(rho1=rho1, rho2=rho2, g=g, C=C)
print("Среднее число N обслуживаемых запросов: {}".format(N))
```

```python
E1_array = []
E2_array = []
lambda1_array = np.arange(lambda1+1)
for lambda1 in lambda1_array:
    rho1 = lambda1/mu
    # rho2 = lambda2/mu # unchange
    # compute E1 and add to array E1_array
    E1_array.append(p_C(C=C, rho1=rho1, rho2=rho2, g=g)) # new rho1 computed in loop
    # compute E2 and add to array E2_array
    E2_array.append(sum_pngc(rho1=rho1, rho2=rho2, g=g, C=C))

import matplotlib.pyplot as plt

plt.plot(lambda1_array, E1_array, color= 'orangered', label="$E_1$")
plt.plot(lambda1_array, E2_array, '',color= 'blue', label="$E_2$")
plt.legend()
plt.title('Вероятность блокировки по времени в зависимости от интенсивности запросов')
plt.xlabel('Интенсивности поступления запросов')
plt.ylabel('Значение вероятности')
plt.show()
```
