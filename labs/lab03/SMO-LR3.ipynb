```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def factorial(i):
    factorial = 1
    for j in range(2,i+1):
        factorial = factorial*j;
    return factorial
```

```python
C = 15 #количество приборов
lambda1 = [] #массив для Lambda_1
lambda2 = 10
start_value = 5 #Начальное Лямбда 1
end_value = 30 #Конечное Лямбда 1
theta1=0.4 
theta2 = 0.5
n1 = 10
n2 = 10
a2 = lambda2*theta1/C
a1 = []
p = [[]]
step = 0.1
step = 0.2 #шаг для Lambda_1

while(start_value <= end_value):
    start_value += step
    lambda1.append(start_value)

##for i in range(0, (len(lambda1))):
    #a1.append(lambda1[i]*theta1/C)
    #N1.append(0)
   #N2.append(0)
    #T1.append(0)
    #T2.append(0)
```

```python
p= np.zeros((n1+1, n2+1), dtype=float).tolist()
lambda1.append(start_value)
for i in range(0, len(lambda1)):
    a1.append(lambda1[i]*theta1/C)
    for n1 in range(0, n1+1):
        for n2 in range(0, n2+1):
            p[0][0] = p[0][0] + factorial(n1+n2)*(a1[i]**n1)*(a2**n2)/(factorial(n1)*factorial(n2))
               
    p[0][0]=1/p[0][0]

    for n1 in range(0, n1+1):
        for n2 in range(0, n2+1):
            if (n1==0 and n2==0):
                continue
            else:
                p[n1][n2]=p[0][0]*factorial(n1+n2)*(a1[i]**n1)*(a2**n2)/(factorial(n1)*factorial(n2))
print('Распределение вероятностей для каждого состояний:\n', p)
    
```

```python
s = 0
for v in p:
    s = s+sum(v)
print("Сумма: ", s)
```

```python
N1 = []
T1 = []
N2 = []
T2 = []
for i in range(0, (len(lambda1))):
    N1.append(0)
    N2.append(0)
    T1.append(0)
    T2.append(0)
    N1[i]=lambda1[i]*theta1/(lambda1[i]*theta1+lambda2*theta2)
    N2[i]=lambda2*theta2/(lambda1[i]*theta1+lambda2*theta2)
    
    T1[i]=N1[i]/lambda1[i]
    T2[i]=N2[i]/lambda2
fig = plt.figure(figsize=(10,10))
plt.plot(lambda1, N1, label = "$N_1$")
plt.plot(lambda1, N2, label = "$N_2$")
plt.title("Среднее число обслуживаемых в системе запросов")
plt.xlabel("Lambda_1")
plt.ylabel("Amount")
plt.legend()
```

```python
fig = plt.figure(figsize=(10,10))
plt.plot(lambda1, T1, label = "$T_1$")
plt.plot(lambda1, T2, label = "$T_2$")
plt.title("Среднее время обслуживания запроса")
plt.xlabel("Lambda_1")
plt.ylabel("Time")
plt.legend()
```
