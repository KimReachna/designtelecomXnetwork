–ö–æ–≥–¥–∞ —Ç–µ–∫—É—â–µ–µ –æ–±–Ω–æ–≤–ª–µ–Ω–∏–µ –¥–æ—Å—Ç–∏–≥–∞–µ—Ç $k$ —Å–∞–º—ã—Ö —Ä–∞–Ω–Ω–∏—Ö —É–∑–ª–æ–≤, –∏—Å—Ö–æ–¥–Ω—ã–π —É–∑–µ–ª –∑–∞–≤–µ—Ä—à–∞–µ—Ç –æ—Å—Ç–∞–≤—à–∏–µ—Å—è $n ‚àí k$ –ø–µ—Ä–µ–¥–∞—á –∏ –Ω–∞—á–∏–Ω–∞–µ—Ç –∂–¥–∞—Ç—å —Å–ª–µ–¥—É—é—â–µ–≥–æ –ø—Ä–∏–±—ã—Ç–∏—è, –∞ –∑–∞—Ç–µ–º –ø–æ–≤—Ç–æ—Ä—è–µ—Ç –ø—Ä–æ—Ü–µ—Å—Å, –∫–æ–≥–¥–∞ –ø—Ä–∏–±—ã–≤–∞–µ—Ç —Å–ª–µ–¥—É—é—â–∏–π –ø–∞–∫–µ—Ç –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è. –ü–æ—Å–∫–æ–ª—å–∫—É –∑–∞–¥–µ—Ä–∂–∫–∏ –Ω–∞ –ª–∏–Ω–∏—è—Ö —è–≤–ª—è—é—Ç—Å—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏ –∏ –æ–¥–∏–Ω–∞–∫–æ–≤–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–º–∏ i.i.d., –∫–æ–Ω–µ—á–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –ø–æ–ª—É—á–∞—é—Ç –ø–∞–∫–µ—Ç –≤ –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏–∏ —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é $p = \frac{k}{n}. –ï—Å–ª–∏ –∫–æ–Ω–µ—á–Ω—ã–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –ø–æ–ª—É—á–∞–µ—Ç –æ–±–Ω–æ–≤–ª–µ–Ω–∏–µ $j$ –∏ —Å–ª–µ–¥—É—é—â–µ–µ, –∫–æ—Ç–æ—Ä–æ–µ –æ–Ω –ø–æ–ª—É—á–∞–µ—Ç, $n$, —è–≤–ª—è–µ—Ç—Å—è –æ–±–Ω–æ–≤–ª–µ–Ω–∏–µ–º $j + M$, —Ç–æ $M$ —è–≤–ª—è–µ—Ç—Å—è –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–º —Å $p$ —Å –º–æ–º–µ–Ω—Ç–∞–º–∏:

\end{equation}

–ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ [20], —Å—Ä–µ–¥–Ω–∏–π –≤–æ–∑—Ä–∞—Å—Ç –¥–ª—è —Å–∞–º–æ–π —Ä–∞–Ω–Ω–µ–π —Å—Ö–µ–º—ã –æ—Å—Ç–∞–Ω–æ–≤–∫–∏ $k$ —Å —ç–∫–∑–æ–≥–µ–Ω–Ω—ã–º–∏ –ø–æ—Å—Ç—É–ø–ª–µ–Ω–∏—è–º–∏ –ø–∞–∫–µ—Ç–æ–≤ —Å $Œº$ —Ä–∞–≤–µ–Ω:

\begin{equation}
\tag{6}

\end{equation}
